## Delay time
kinetic energy:
$$T = E - m_0\cdot c^2$$
$$T = m_0\cdot c^2(\gamma -1),\qquad  \gamma = \frac{1}{\sqrt{1-\beta^2}}$$
delay time between massive and massless particles:
$$\Delta t = \frac{L}{c}\cdot\Big{(}\frac{1}{\beta}-1\Big{)}$$
Let's denote:
$$\alpha = \frac{T}{m_0\cdot c^2}$$
then: 
$$\beta = \frac{\sqrt{\alpha^2+2\alpha}}{\alpha+1}$$
and finally
$$\Delta t = \frac{L}{c}\cdot\Big{(}\frac{\alpha+1}{\sqrt{\alpha^2+2\alpha}}-1\Big{)}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

c = 3e8
m0 = 1e-6 # neutrino mass in MeV units
L = np.arange(1000, 1e8+500, 1000)
T = np.arange(1e-1, 10.1, 0.1)

L, T = np.meshgrid(L,T)
alpha = T / m0
delta_t = L / c *((alpha + 1) / np.sqrt(alpha ** 2 + 2 * alpha) - 1)

# plot figure
fig = plt.figure(figsize=[13, 6], dpi=140)
cs = plt.contourf(L, T, delta_t, cmap=cm.gnuplot)
cbar = plt.colorbar(cs)
plt.title('mass = 1 eV')
plt.show()
```

```python
def delta_t_func(L=2000, T=1):
    global m0
    alpha = T / m0
    delta_t = L / 3e8 *((alpha + 1) / np.sqrt(alpha ** 2 + 2 * alpha) - 1)
    return delta_t

print(delta_t_func())
```
